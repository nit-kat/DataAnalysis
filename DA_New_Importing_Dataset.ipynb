```python
import numpy as np
import pandas as pd
```

```python
pathh = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(pathh,header=None)
df.head(3)
```

```python
df.tail(10)
```

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

df.columns = headers
df.head(3)
```

```python
df1 = df.replace('?',np.NaN)

df = df1.dropna(subset=['price'],axis=0)

df.head(10)
```

```python
# Find the name of the columns of the dataframe
df.columns
```

```python
df.to_csv("automobilesdata.csv",index=False)
```
